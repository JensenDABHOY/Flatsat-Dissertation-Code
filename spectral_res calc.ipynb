```python
l = [300, 315, 330] #in nm
l_plus_dl = [300.1, 315.1, 330.1] #in nm
delta_l = 0.1
RMS_y_radii_l = [0.00629, 0.006473, 0.007094] #in mm
RMS_y_radii_l_plus_dl = [0.00629, 0.006473, 0.007094] #in mm
RMS_y_radii_l_2 = [0.01068, 0.01023, 0.00975] #in mm
RMS_y_radii_l_plus_dl_2 = [0.001068, 0.01023, 0.009746] #in mm
y_bases = [-19.103, -21.287, -23.460] #in mm
y_offsets = [-19.107, -21.291, -23.464] #in mm
y_bases_2 = [-15.094, -15.854, -16.610] #in mm
y_offsets_2 = [-15.099, -15.859, -16.615] #in mm
deltays = []
dy_by_dls = []
FWHMs = []
Rs = []
deltays_2 = []
dy_by_dls_2 = []
FWHMs_2 = []
Rs_2 = []
delta_lambda_resolutions = []
delta_lambda_resolutions_2 = []
for i in [0,1,2]:
    deltay = y_bases[i] - y_offsets[i]
    deltays.append(deltay)
    dy_by_dl = deltay/delta_l
    dy_by_dls.append(dy_by_dl)
    FWHM = RMS_y_radii_l[i] * 2.355
    FWHMs.append(FWHM)
    delta_lambda_resolution = FWHM/dy_by_dl
    delta_lambda_resolutions.append(delta_lambda_resolution) #in nm
    R = l[i]/delta_lambda_resolution
    Rs.append(R)

    deltay_2 = y_bases_2[i] - y_offsets_2[i]
    deltays_2.append(deltay_2)
    dy_by_dl_2 = deltay_2/delta_l
    dy_by_dls_2.append(dy_by_dl_2)
    FWHM_2 = RMS_y_radii_l_2[i] * 2.355
    FWHMs_2.append(FWHM_2)
    delta_lambda_resolution_2 = FWHM_2/dy_by_dl_2
    delta_lambda_resolutions_2.append(delta_lambda_resolution_2) #in nm
    R_2 = l[i]/delta_lambda_resolution_2
    Rs_2.append(R_2)
```
